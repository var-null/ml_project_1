```python
import pandas as pd
import numpy as np
import plotly.express as px


df = pd.read_csv("tips.csv")
df.head()

```

```python
df['tip_percent'] = df['tip'] / df['total_bill'] * 100
print(df[['total_bill','tip','tip_percent']].head())
```

```python
fig = px.pie(df, names="sex", title="Distribution by Sex")
fig.show()
```

```python
import plotly.express as px

fig = px.scatter(df, x='total_bill', y='tip', color='sex',
                 size='size', hover_data=['day', 'time'],
                 title='Total Bill vs Tip by Sex')
fig.show()
```
