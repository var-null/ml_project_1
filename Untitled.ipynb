```python
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_csv("tips.csv")

df.head()

```

```python
fig = px.pie(df, names="sex", title="Distribution by Sex")
fig.show()
```

```python

fig = px.scatter(df, x='total_bill', y='tip', color='sex',
                 size='size', hover_data=['day', 'time'],
                 title='Total Bill vs Tip by Sex')
fig.show()
```

```python


df['sex'] = df['sex'].map({'Male': 1, 'Female': 0})
df['smoker'] = df['smoker'].map({'Yes': 1, 'No': 0})
df['time'] = df['time'].map({'Dinner': 1, 'Lunch': 0})

df = pd.get_dummies(df, columns=['day'], drop_first=True)

df.head()
```
